```python
rainfall = pd.read_csv("https://raw.githubusercontent.com/SoongMoo/soldesk20250412/refs/heads/main/data/Seattle2014.csv")
rainfall # DataFrame
inches = np.array(rainfall['PRCP'] / 254) 
inches
# 0.5인치이상 온 날은 몇일입니까?
#True :1 , False :0
print(np.sum(inches > 0.5))
# 강수량이 0.5에서 0.8 사이인 날은 몇일이나 되나요?
print(np.sum((inches > 0.5) & (inches < 0.8))) # & : and
#강수량이 0.5보다 작고 0.8이 온날은?
print(np.sum(~((inches > 0.5) & (inches < 0.8))))
print(np.sum((inches <= 0.5) | (inches >= 0.8))) # | : or
#비가 안온 날
print(np.sum(inches == 0))
#비가 온 날
print(np.sum(inches != 0))
```
